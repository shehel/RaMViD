```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
all_images = np.load('/tmp/openai-2022-11-18-20-22-14-612490/true.npz')    
all_images = all_images['arr_0']

imgs = []
fig, ax = plt.subplots(figsize=(8, 8))
for img in (all_images[0, 0,:,:,:]):
    img = ax.imshow(img, animated=True)
    imgs.append([img])

ani = animation.ArtistAnimation(fig, imgs, interval=1000, blit=True, repeat_delay=2000)
HTML(ani.to_html5_video())
```
